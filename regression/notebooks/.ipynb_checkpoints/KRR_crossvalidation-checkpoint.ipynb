```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
from scipy import interpolate
from netCDF4 import Dataset

import utils as utl
```

## Define constants

```python
Nh = 96
Nt = 37
sspacing = 3
tspacing = 4

HTLS_sknots = np.arange(0,Nh,sspacing)
HTHS_sknots = np.arange(0,Nh,1)
LTHS_tknots = np.arange(0,Nh,tspacing)
Nl = len(HTLS_sknots)
Ns = len(LTHS_tknots)

Do = Nh*Nh
Di = Nl*Nl

N = Nt*Ns
```

## Load sample snapshots

```python
ncfile1 = Dataset('/data/ISOTROPIC/data/data_downsampled4.nc','r')
xh = np.array(ncfile1.variables['velocity_x'][0,0:Nh,0:Nh,0])
ncfile1.close()
xl = xh[0:-1:sspacing,0:-1:sspacing]
xinterp = utl.interp2(HTLS_sknots, HTLS_sknots, xl, HTHS_sknots, HTHS_sknots)

fig = plt.figure(figsize=(13, 5))

ax = plt.subplot(1,3, 1)
cax = plt.imshow(xh, interpolation='none',clim=[-3,3],cmap='jet')
ax.set_xticks(())
ax.set_yticks(())
    
ax = plt.subplot(1,3, 2)
cax = plt.imshow(xl, interpolation='none',clim=[-3,3],cmap='jet')
ax.set_xticks(())
ax.set_yticks(())

ax = plt.subplot(1,3, 3)
cax = plt.imshow(xinterp, interpolation='none',clim=[-3,3],cmap='jet')
ax.set_xticks(())
ax.set_yticks(())

# fig.colorbar(cax, ticks=[-3, -2, -1, 0, 1, 2, 3])
```

## Load all input/output data and centerize

```python
Xo_tr = np.zeros((N, Do))
Xi_tr = np.zeros((N, Di))
ncfile1 = Dataset('/data/ISOTROPIC/data/data_downsampled4.nc','r')
for t in range(Nt):
    count = 0
    for i in LTHS_tknots:
        xh = np.array(ncfile1.variables['velocity_x'][t,0:Nh,0:Nh,i])
        xl = xh[0:-1:sspacing,0:-1:sspacing] # xh[np.meshgrid(HTLS_sknots,HTLS_sknots)]
        Xo_tr[t*Ns + count,:] = np.reshape(xh,(1, Do))
        Xi_tr[t*Ns + count,:] = np.reshape(xl,(1, Di))
        count = count + 1
ncfile1.close()

# normalized: centered, variance 1
mea_i = np.zeros(Di)
sig_i = np.zeros(Di)
for k in range(Di):
    mea_i[k] = np.mean(Xi_tr[:,k])
    sig_i[k] = np.std(Xi_tr[:,k])
    Xi_tr[:,k] = (Xi_tr[:,k]-mea_i[k])/sig_i[k] 
    
mea_o = np.zeros(Do)
sig_o = np.zeros(Do)
for k in range(Do):
    mea_o[k] = np.mean(Xo_tr[:,k])
    sig_o[k] = np.std(Xo_tr[:,k])
    Xo_tr[:,k] = (Xo_tr[:,k]-mea_o[k])/sig_o[k] 
```

```python
fig = plt.figure(figsize=(15, 4))
ax=plt.plot(Xi_tr[:,1])
```

## Use GridSearchCV to get a rough idea of parameters

```python
from sklearn.grid_search import GridSearchCV
from sklearn.kernel_ridge import KernelRidge

kr = GridSearchCV(KernelRidge(kernel='rbf'), cv=6,
                  param_grid={"alpha": np.logspace(-5, -1, 10), "gamma": np.logspace(-10, 0, 5)})
kr.fit(Xi_tr, Xo_tr)

print kr.best_estimator_
```

## Cross-validation curve for gamma

```python
from sklearn.learning_curve import validation_curve
param_range=np.logspace(-8, -3, 20)
train_scores_gamma, test_scores_gamma = validation_curve(KernelRidge(kernel='rbf', alpha=kr.best_estimator_.alpha), 
                                                         Xi_tr, Xi_tr, param_name="gamma", param_range=param_range, 
                                                         scoring = "mean_squared_error", cv=10)
```

```python
# API always tries to maximize a loss function, so MSE is actually in the flipped sign
train_scores_mean_gamma = np.mean(-train_scores_gamma, axis=1)
train_scores_std_gamma = np.std(-train_scores_gamma, axis=1)
test_scores_mean_gamma = np.mean(-test_scores_gamma, axis=1)
test_scores_std_gamma = np.std(-test_scores_gamma, axis=1)

plt.figure(figsize=(7, 5))
plt.title("Validation Curve with KRR")
plt.xlabel("$gamma$")
plt.ylabel("MSE")
# plt.ylim(0, 1.1)
plt.semilogx(param_range, train_scores_mean_gamma, label="Training score", color="r")
plt.fill_between(param_range, train_scores_mean_gamma - train_scores_std_gamma, 
                 train_scores_mean_gamma + train_scores_std_gamma, alpha=0.2, color="r")
plt.semilogx(param_range, test_scores_mean_gamma, label="Cross-validation score", color="g")
plt.fill_between(param_range, test_scores_mean_gamma - test_scores_std_gamma, 
                 test_scores_mean_gamma + test_scores_std_gamma, alpha=0.2, color="g")
plt.legend(loc="best")
plt.show()
```

## Cross-validation curve for alpha

```python
from sklearn.learning_curve import validation_curve
param_range = np.append(0, np.logspace(-10, -1, 21))
train_scores_alpha, test_scores_alpha = validation_curve(KernelRidge(kernel='rbf', gamma=kr.best_estimator_.gamma), 
                                                         Xi_tr, Xi_tr, param_name="alpha", param_range=param_range, 
                                                         scoring = "mean_squared_error", cv=10)

# API always tries to maximize a loss function, so MSE is actually in the flipped sign
train_scores_mean_alpha = np.mean(-train_scores_alpha, axis=1)
train_scores_std_alpha = np.std(-train_scores_alpha, axis=1)
test_scores_mean_alpha = np.mean(-test_scores_alpha, axis=1)
test_scores_std_alpha = np.std(-test_scores_alpha, axis=1)
```

```python
plt.figure(figsize=(7, 5))
plt.title("Validation Curve with KRR")
plt.xlabel("$alpha$")
plt.ylabel("MSE")
# plt.ylim(0, 1.1)
plt.semilogx(param_range, train_scores_mean_alpha, label="Training score", color="r")
plt.fill_between(param_range, train_scores_mean_alpha - train_scores_std_alpha, 
                 train_scores_mean_alpha + train_scores_std_alpha, alpha=0.2, color="r")
plt.semilogx(param_range, test_scores_mean_alpha, label="Cross-validation score", color="g")
plt.fill_between(param_range, test_scores_mean_alpha - test_scores_std_alpha, 
                 test_scores_mean_alpha + test_scores_std_alpha, alpha=0.2, color="g")
#plt.legend(loc="best")
plt.show()
```
